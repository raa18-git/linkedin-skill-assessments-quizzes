```python
import numpy as np
import matplotlib.pyplot as plt
# GIVEN DATA
temperatures = np.array([30, 30, 30, 30, 40, 40, 40, 50, 50, 50, 60, 60, 60, 60])
hardness = np.array([55.8, 59.1, 54.8, 54.6, 43.1, 42.2, 45.2, 31.6, 30.9, 30.8, 17.5, 20.5, 17.2, 16.9])
# PERFORM LINEAR REGRESSION
coefficients = np.polyfit(temperatures, hardness, 1)
slope, intercept = coefficients
# COMPUTE THE LINEAR HYPOTHESIS
linear_hypothesis = slope * temperatures + intercept
# CALCULTAE R- SQUARED
residuals = hardness - linear_hypothesis
total_sum_of_squares = np.sum((hardness - np.mean(hardness)) ** 2)
residual_sum_of_squares = np.sum(residuals ** 2)
r_squared = 1 - (residual_sum_of_squares / total_sum_of_squares)
# PLOT DATA
plt.scatter(temperatures, hardness, label='Data')
plt.plot(temperatures, linear_hypothesis, color='red', label='Linear Hypothesis')
plt.xlabel('Test Temperature (°C)')
plt.ylabel('Hardness')
plt.title('Effect of Temperature on Hardness')
plt.legend()
plt.grid()
# PRINT THE COFF AND R SQUARED
print(f"Slope: {slope:.4f}")
print(f"Intercept: {intercept:.4f}")
print(f"R-squared: {r_squared:.4f}")
# SHOW THE PLOT
plt.show()
```
